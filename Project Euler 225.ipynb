```python
def inDict(dict, k1, k2, k3):
    if (k1 not in dict):
        return False
    if (k2 not in dict[k1]):
        return False
    return (k3 in dict[k1][k2])

def addToDict(dict, k1, k2, k3, value):
    if (k1 not in dict):
        dict[k1] = {}
    if (k2 not in dict[k1]):
        dict[k1][k2] = {}
    dict[k1][k2][k3] = value

def divides(n):
    Tn = 1
    Tn1 = 1
    Tn2 = 1
    dict = {}

    while (Tn != 0 and not inDict(dict, Tn, Tn1, Tn2)):
        addToDict(dict, Tn, Tn1, Tn2, True)
        temp = Tn
        Tn += Tn1 + Tn2
        Tn %= n
        Tn2 = Tn1
        Tn1 = temp
    return (Tn == 0)

n = 25
count = 0
while (count < 124):
    if (not divides(n)):
        count += 1
        print(n)
    n += 2
```
