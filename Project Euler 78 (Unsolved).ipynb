```python
def addToDictionary(dictionary, key1, key2, value):
    if key1 not in dictionary:
        dictionary[key1] = {}
    dictionary[key1][key2] = value

def p(n, dictionary):
    return getNumWaysToPile(n, n, dictionary)
    
def getNumWaysToPile(numCoins, maxPileSize, dictionary):
    if (maxPileSize > numCoins):
        maxPileSize = numCoins
    if numCoins in dictionary:
        if maxPileSize in dictionary[numCoins]:
            return dictionary[numCoins][maxPileSize]
    if (numCoins == 0 or maxPileSize == 1):
        return 1
    if (maxPileSize == 2):
        return (numCoins + 2) // 2
    if (maxPileSize == 0):
        return 0
    count = 0
    for i in range(1, maxPileSize+1):
        count += getNumWaysToPile(numCoins - i, i, dictionary)
#        count %= 100
    addToDictionary(dictionary, numCoins, maxPileSize, count)
    return count

dictionary = {}
p(100, dictionary)

for key in range(100):
    if key in dictionary:
        if 3 in dictionary[key]:
            print ("%d: %d" %(key, dictionary[key][3]))
```
